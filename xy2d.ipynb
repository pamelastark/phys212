```python
fig, ax = plt.subplots(2, 5, figsize=(16, 8))
lattice = XY2D(L=50, beta=3)
#fig.suptitle(fr'Monte Carlo Evolution of States; $\beta$={lattice.beta}', fontsize=14)
for s in range(200):
    lattice.step()
    if s%20==0:
        lattice.plot2D(fig, ax[(s//20)//5][(s//20)%5])  
        ax[(s//20)//5][(s//20)%5].set_xlabel(f'{20 * s//20} steps')

plt.savefig('state_evolution.jpg', bbox_inches='tight')
#lattice.show('equilibrium_beta05_3000.jpg')
```

```python
config_matrix = np.reshape(lattice.theta, (lattice.L, lattice.L))
X, Y = np.meshgrid(np.arange(0, lattice.L), np.arange(0, lattice.L))
U = np.cos(config_matrix)
V = np.sin(config_matrix)
```

```python
plt.figure(figsize=(4, 4), dpi=100)
# plt.title('Arrows scale with plot width, not view')
Q = plt.quiver(X, Y, U, V, units='width')
qk = plt.quiverkey(Q, 0.1, 0.1, 1, r'$spin$', labelpos='E',
                   coordinates='figure')
plt.axis('off')
plt.show()
```
