```python
import sys
sys.setrecursionlimit(100000)
from XY3D import *
import time
```

```python
%timeit
fig, ax = plt.subplots(2, 5, figsize=(16, 8))
lattice = XY3D(L=20, beta=1000)

for s in range(200):
    lattice.step()
    if s%20==0:
        lattice.plot2D(fig, ax[(s//20)//5][(s//20)%5])  
        ax[(s//20)//5][(s//20)%5].set_xlabel(f'{20 * s//20} steps')

plt.savefig('state_evolution3D_L20_b1000.jpg', bbox_inches='tight')
#lattice.show('equilibrium_3D_3000.jpg')
```
